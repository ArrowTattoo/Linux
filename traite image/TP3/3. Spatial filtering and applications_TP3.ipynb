## Mean filter

We propose to filter the _cameraman.png_ image by using a mean filter with this kernel: 

\begin{equation*}
h_1(i,j) = 
\frac{1}{9} \times 
\begin{pmatrix}
1 & 1 & 1 \\
1 & \bf{1} & 1 \\
1 & 1& 1  
\end{pmatrix}
\end{equation*}

**Responses:**


```python

# RAZ (figures closing)
plt.close('all')

# image reading
im = plt.imread('../standard_test_images/cameraman.png')

# ... and normalization (if necessary)
print("Test, type of each elemant:",im.dtype)
print("Test, max of the elemant:", im.max())
im=im/im.max()
im=255*im
im=im.astype(np.uint8)

# apply kernel
kernel1 =np.array([[1,1,1],[1,1,1],[1,1,1]])/9
im1=cv.filter2D(src=im,ddepth=-1,kernel=kernel1)
 
# display 
ig,axs=plt.subplots(1,2)
axs[0].title.set_text('camerman')
axs[0].imshow(im,cmap='gray')
axs[1].title.set_text('filtered')
axs[1].imshow(im1,cmap='gray')
```

## Differentiator: Sharpening a image

We propose to filter the _cameraman.png_ image by using this second kernel: 

\begin{equation*}
h_2(i,j) = 
\begin{pmatrix}
0 & -1 & 0 \\
-1 & \bf{5} & -1 \\
0 & -1 & 0  
\end{pmatrix}
\end{equation*}

In your python code,  load of the image, filter the image, and display the images (original and filtered)

**Questions:**

- what happens if the gain of the kernel equals 0? (define the new kernel and experiment it)

**Responses:**


```python
 
# RAZ (figures closing)
plt.close('all')

# image reading
im = plt.imread('../standard_test_images/cameraman.png')

# ... and normalization (if necessary)
print("Test, type of each elemant:",im.dtype)
print("Test, max of the elemant:", im.max())
im=im/im.max()
im=255*im
im=im.astype(np.uint8)

# apply kernel
kernel2 =np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
im2=cv.filter2D(src=im,ddepth=-1,kernel=kernel2)
 
# display    
ig,axs=plt.subplots(1,2)
axs[0].title.set_text('camerman')
axs[0].imshow(im,cmap='gray')
axs[1].title.set_text('filtered')
axs[1].imshow(im1,cmap='gray')

plt.figure()
plt.title('camerman-gradients')
plt.imshow(im2,cmap='jet')
plt.colorbar()
plt.show()
```
