```
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy.optimize import linprog # Le moteur pour HiGHS

# --- 1. PR√âPARATION DES POINTS ---
m = 100
ec_b = 150
x = np.linspace(0, 1, m)

# --- 2. COURBE FIXE ---
a, b, c = 6500, -6000, 1500
y = a*x**2 + b*x + c
y_bruite = y + np.random.normal(0, ec_b, m)

# --- R√âSOLUTION NORME 1 (L1) AVEC HiGHS ---
# Variables : [a_sol, b_sol, c_sol, t1, t2, ..., tm] -> total 3 + m variables
n_vars = 3 + m

# Objectif : minimiser la somme des t_i (indices 3 √† la fin)
c_obj = np.zeros(n_vars)
c_obj[3:] = 1

# Contraintes : |y_i - (ax^2 + bx + c)| <= t_i
# Cela donne deux lignes par point i :
# 1)  ax^2 + bx + c - t_i <= y_i
# 2) -ax^2 - bx - c - t_i <= -y_i
A_ub = []
b_ub = []

for i in range(m):
    xi2, xi = x[i]**2, x[i]
    # Ligne 1
    row1 = np.zeros(n_vars)
    row1[0:3] = [xi2, xi, 1]
    row1[3+i] = -1
    A_ub.append(row1)
    b_ub.append(y_bruite[i])

    # Ligne 2
    row2 = np.zeros(n_vars)
    row2[0:3] = [-xi2, -xi, -1]
    row2[3+i] = -1
    A_ub.append(row2)
    b_ub.append(-y_bruite[i])

# On laisse les variables libres (peuvent √™tre n√©gatives)
res = linprog(c_obj, A_ub=A_ub, b_ub=b_ub, method='highs')
a_sol, b_sol, c_sol = res.x[0:3]

# --- 3. COURBE AL√âATOIRE ---
a_rand = random.randint(7000, 10000)
b_rand = random.randint(-7000, -4000)
c_rand = random.randint(1600, 3000)

y_rand = a_rand*x**2 + b_rand*x + c_rand
y_rand_bruite = y_rand + np.random.normal(0, ec_b, m)

# --- 4. AFFICHAGE ET DESSIN ---
plt.plot(x, y, color="blue", label="Th√©orique Fixe", linewidth=2)
plt.scatter(x, y_bruite, color="blue", s=15, alpha=0.5, label="Points Bruit√©s")

# On dessine la courbe trouv√©e par HiGHS
y_recupere = a_sol*x**2 + b_sol*x + c_sol
plt.plot(x, y_recupere, color="green", linestyle="--", label=f"L1 (HiGHS): a={a_sol:.1f}")

print(f"Coefficients trouv√©s par Norme 1 : a={a_sol:.2f}, b={b_sol:.2f}, c={c_sol:.2f}")
print(f"√âcart Norme 2 avec le vrai 'a' : {abs(a - a_sol):.2f}")

plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```
import matplotlib.pyplot as plt  # Outil pour dessiner le graphique
import numpy as np               # Outil pour cr√©er les listes de chiffres
import random                    # Outil pour g√©n√©rer du hasard

# --- 1. PR√âPARATION DES POINTS ---
m = 100
ec_b = 150 # ecart du bruit √† g√©n√©rer
x = np.linspace(0, 1, m)       # 100 points entre 0 et 1

# --- CONFIGURATION DU GRAPHIQUE (C√îTE √Ä C√îTE) ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- 2. COURBE FIXE (Sur le premier graphique) ---
a, b, c = 6500, -6000, 1500
y = a*x**2 + b*x + c
bruit_fixe = np.random.normal(0, ec_b, m)
y_bruite = y + bruit_fixe

ax1.plot(x, y, color="blue", label=f"Fixe: a={a}, b={b}, c={c}")
ax1.scatter(x, y_bruite, color="blue", s=10, alpha=0.7)

# --- 3. COURBE AL√âATOIRE (Sur le premier graphique) ---
a_rand = random.randint(7000, 10000)
b_rand = random.randint(-7000, -4000)
c_rand = random.randint(1600, 3000)

y_rand = a_rand*x**2 + b_rand*x + c_rand
bruit_rand = np.random.normal(0, ec_b, m)
y_rand_bruite = y_rand + bruit_rand

ax1.plot(x, y_rand, color="red", label=f"Aleat (int): a*={a_rand}")
ax1.scatter(x, y_rand_bruite, color="red", s=10, alpha=0.7, marker="*")
ax1.set_title("Loi Normale (Entiers)")
ax1.legend(fontsize='small')
ax1.grid(linestyle='--', alpha=0.5)

# --- 4. NOUVELLE COURBE : LOI UNIFORME (Sur le second graphique) ---
# Utilisation de nombres √† virgule pour les coefficients
a_uni = random.uniform(7000, 10000)
b_uni = random.uniform(-7000, -4000)
c_uni = random.uniform(1600, 3000)

y_uni = a_uni*x**2 + b_uni*x + c_uni
# Bruit g√©n√©r√© avec une loi uniforme (entre -ec_b et +ec_b)
bruit_uniforme = np.random.uniform(-ec_b, ec_b, m)
y_uni_bruite = y_uni + bruit_uniforme

ax2.plot(x, y_uni, color="green", label=f"Uniforme: a={a_uni:.1f}")
ax2.scatter(x, y_uni_bruite, color="green", s=10, alpha=0.7, marker="x")
ax2.set_title("Loi Uniforme (D√©cimaux)")
ax2.legend(fontsize='small')
ax2.grid(linestyle='--', alpha=0.5)

# --- 5. AFFICHAGE DES VALEURS ---
print("--- Valeurs Al√©atoires (randint) ---")
print(f"a = {a_rand}, b = {b_rand}, c = {c_rand}")
print(f"Norme 2 vs Fixe = {((a_rand-a)**2 + (b_rand-b)**2 + (c_rand-c)**2)**0.5:.2f}")

print("\n--- Valeurs Uniformes (float) ---")
print(f"a_uni = {a_uni:.2f}, b_uni = {b_uni:.2f}, c_uni = {c_uni:.2f}")

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt  # Importation pour cr√©er le graphique
import numpy as np               # Importation pour les calculs de listes et bruits
import random                    # Importation pour les tirages al√©atoires

# ------- 1. VALEURS TH√âORIQUES -----------------
a, b, c = 6500, -6000, 1500      # Param√®tres de la courbe fixe (bleue)
m = 100                          # Nombre de points (√©chantillons)
ec_b = 150                       # Intensit√© du bruit (√©cart-type ou borne)
x = np.linspace(0, 1, m)         # 100 points de 0 √† 1 pour l'axe X
y = a*x**2 + b*x + c             # √âquation de la courbe th√©orique

# ------- 2. VALEURS AL√âATOIRES (COEFFICIENTS) -------------
# On g√©n√®re les coefficients une seule fois pour les deux graphes ou s√©par√©ment
a_alea = random.randint(7000, 10000)
b_alea = random.randint(-7000, -4000)
c_alea = random.randint(1600, 3000)
y_alea = a_alea*x**2 + b_alea*x + c_alea # Courbe avec coefficients al√©atoires

# Calcul de la distance entre la courbe th√©orique et l'al√©atoire (Norme 2)
norme_alea = ((a_alea-a)**2 + (b_alea-b)**2 + (c_alea-c)**2)**0.5

# -------------- 3. AFFICHAGE DE GRAPHES C√îTE √Ä C√îTE -----------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# --- GRAPHE 1 : LOI NORMALE ---
# G√©n√©ration du bruit normal
bruit_normal_fixe = np.random.normal(0, ec_b, m)
y_bruit_normal_fixe = y + bruit_normal_fixe

bruit_normal_alea = np.random.normal(0, ec_b, m)
y_alea_br_norm = y_alea + bruit_normal_alea

# Dessin Graphe 1
ax1.plot(x, y, color="blue", label=f"Fixe: a={a}, b={b}, c={c}") # Ligne th√©orique
ax1.scatter(x, y_bruit_normal_fixe, color="blue", s=10, alpha=0.4) # Points th√©oriques bruit√©s
ax1.plot(x, y_alea, color="red", label=f"Aleat: a*={a_alea}, b*={b_alea}, c*={c_alea}") # Ligne al√©atoire
ax1.scatter(x, y_alea_br_norm, color="red", s=10, alpha=0.7, marker="*") # Points al√©atoires bruit√©s

ax1.set_title(f"Bruit Loi Normale\nNorme: {norme_alea:.2f}") # Titre
ax1.legend(fontsize='x-small') # L√©gende
ax1.grid(linestyle='--', alpha=0.5) # Quadrillage

# --- GRAPHE 2 : LOI UNIFORME ---
# G√©n√©ration du bruit uniforme
bruit_uni_fixe = np.random.uniform(-ec_b, ec_b, m)
y_bruit_uni_fixe = y + bruit_uni_fixe

bruit_uni_alea = np.random.uniform(-ec_b, ec_b, m)
y_alea_br_uni = y_alea + bruit_uni_alea

# Dessin Graphe 2
ax2.plot(x, y, color="blue", label=f"Fixe: a={a}, b={b}, c={c}") # Ligne th√©orique
ax2.scatter(x, y_bruit_uni_fixe, color="blue", s=10, alpha=0.4) # Points th√©oriques bruit√©s
ax2.plot(x, y_alea, color="green", label=f"Aleat: a*={a_alea}, b*={b_alea}, c*={c_alea}") # Ligne al√©atoire
ax2.scatter(x, y_alea_br_uni, color="green", s=10, alpha=0.7, marker="x") # Points al√©atoires bruit√©s

ax2.set_title(f"Bruit Loi Uniforme\nNorme: {norme_alea:.2f}") # Titre
ax2.legend(fontsize='x-small') # L√©gende
ax2.grid(linestyle='--', alpha=0.5) # Quadrillage

# --- 4. AFFICHAGE CONSOLE ---
print("--- R√âSULTATS G√âN√âRAUX ---")
print(f"Valeurs fix√©es  : a={a}, b={b}, c={c}")
print(f"Valeurs al√©at   : a*={a_alea}, b*={b_alea}, c*={c_alea}")
print(f"Norme Euclidienne ||(a*,b*,c*) - (a,b,c)||2 = {norme_alea:.2f}")

plt.tight_layout() # √âvite les chevauchements
plt.show() # Affiche le r√©sultat final
```

```
import matplotlib.pyplot as plt  # Importe l'outil pour dessiner les graphiques
import numpy as np               # Importe l'outil pour les calculs math√©matiques et les tableaux
import random                    # Importe l'outil pour g√©n√©rer des nombres al√©atoires





# ------- 1. VALEURS TH√âORIQUES (LA R√âF√âRENCE) -----------------
a, b, c = 6500, -6000, 1500      # D√©finit les coefficients de la courbe "v√©rit√©" (bleue)
m = 100                          # D√©finit le nombre total de points (√©chantillons)
ec_b = 150                       # D√©finit l'intensit√© du bruit (√©cart-type pour Normale, borne pour Uniforme)
x = np.linspace(0, 1, m)         # Cr√©e 100 points espac√©s r√©guli√®rement entre 0 et 1 pour l'axe X
y = a*x**2 + b*x + c             # Calcule les valeurs Y th√©oriques exactes (parabole parfaite)
# ------- 1. VALEURS TH√âORIQUES (LA R√âF√âRENCE) -----------------





# ------- 2. G√âN√âRATION DES COEFFICIENTS AL√âATOIRES (UNE SEULE FOIS) -------------
a_alea = random.randint(7000, 10000)   # Choisit un coefficient 'a' au hasard entre 7000 et 10000
b_alea = random.randint(-7000, -4000)  # Choisit un coefficient 'b' au hasard entre -7000 et -4000
c_alea = random.randint(1600, 3000)    # Choisit un coefficient 'c' au hasard entre 1600 et 3000
y_alea = a_alea*x**2 + b_alea*x + c_alea # Calcule la courbe al√©atoire "cible" avec ces coefficients
# ------- 2. G√âN√âRATION DES COEFFICIENTS AL√âATOIRES (UNE SEULE FOIS) -------------




# ---------------Distance entre param√®tre th√©orique et al√©atoires-------------------
norme_alea = ((a_alea-a)**2 + (b_alea-b)**2 + (c_alea-c)**2)**0.5
# ---------------Distance entre param√®tre th√©orique et al√©atoires-



# -------------- 3. PR√âPARATION DE L'AFFICHAGE C√îTE √Ä C√îTE -----------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8)) # C√¥te √† c√¥te
# -------------- 3. PR√âPARATION DE L'AFFICHAGE C√îTE √Ä C√îTE -----------






# --------------------------- LOI NORMALE (GAUSSIENNE) -------------
         #------Bruit √† ajouter aux points th√©oriques-----
bruit_n_fixe = np.random.normal(0, ec_b, m)       # G√©n√®re un bruit en forme de cloche pour la courbe bleue
y_br_n_fixe = y + bruit_n_fixe

         #------Bruit √† ajouter aux points al√©atoires--------------
bruit_n_alea = np.random.normal(0, ec_b, m)       # G√©n√®re un bruit en forme de cloche pour la courbe al√©atoire
y_br_n_alea = y_alea + bruit_n_alea               # Ajoute ce bruit aux points al√©atoires

ax1.plot(x, y, color="blue", label=f"Fixe: a={a}, b={b}, c={c}") # Dessine la ligne th√©orique bleue
ax1.scatter(x, y_br_n_fixe, color="blue", s=10, alpha=0.4)       # Dessine les points bleus (bruit normal)
ax1.plot(x, y_alea, color="red", label=f"Aleat: a*={a_alea}, b*={b_alea}, c*={c_alea}") # Dessine la ligne rouge
ax1.scatter(x, y_br_n_alea, color="red", s=10, alpha=0.7, marker="*") # Dessine les points rouges (√©toiles)
ax1.set_title(f"Bruit Loi Normale\nNorme: {norme_alea:.2f}")      # D√©finit le titre avec la valeur de la norme
ax1.grid(linestyle='--', alpha=0.5)                               # Ajoute un quadrillage discret
ax1.legend(fontsize='x-small')                                    # Affiche la l√©gende explicative
# -------------------------- LOI NORMALE (GAUSSIENNE) -------------







# ------------------------LOI UNIFORME (RECTANGULAIRE) --------------------------------
          #---ajout du bruit aux points th√©oriques-----------
bruit_u_fixe = np.random.uniform(-ec_b, ec_b, m)  # G√©n√®re un bruit plat (entre -150 et 150) pour la bleue
y_br_u_fixe = y + bruit_u_fixe                    # Ajoute ce bruit uniforme aux points th√©oriques

          #-----ajout de bruit aux points al√©atoires
bruit_u_alea = np.random.uniform(-ec_b, ec_b, m)  # G√©n√®re un bruit plat pour la courbe al√©atoire
y_br_u_alea = y_alea + bruit_u_alea               # Ajoute ce bruit uniforme aux points al√©atoires

ax2.plot(x, y, color="blue", label=f"Fixe: a={a}, b={b}, c={c}") # Dessine la ligne th√©orique bleue (rappel)
ax2.scatter(x, y_br_u_fixe, color="blue", s=10, alpha=0.8)       # Taille des points et leur opacit√©
ax2.plot(x, y_alea, color="green", label=f"Aleat: a*={a_alea}, b*={b_alea}, c*={c_alea}") # Dessine la ligne verte

ax2.scatter(x, y_br_u_alea, color="green", s=10, alpha=0.7, marker="x") # Dessine les points verts (croix)
ax2.set_title(f"Bruit Loi Uniforme\nNorme: {norme_alea:.2f}")      # D√©finit le titre avec la m√™me norme
ax2.grid(linestyle='--', alpha=0.5)                             # Ajoute le quadrillage, 0.5---opacit√© du quadrillage
ax2.legend(fontsize='x-small')                                    # Affiche la l√©gende
# ------------------------LOI UNIFORME (RECTANGULAIRE) --------------------------------





# --- -----------------AFFICHAGE DES R√âSULTATS-------------------------------------

print(f"Coefficients Al√©atoires : a*={a_alea}, b*={b_alea}, c*={c_alea}") # Affiche les valeurs de a, b, c
print(f"Norme Euclidienne (Distance) : {norme_alea:.2f}")         # Affiche la distance calcul√©e

#plt.tight_layout() # Optimise l'espace entre les deux graphiques pour √©viter les chevauchements
plt.show()         # Ouvre la fen√™tre et affiche les graphiques finaux
```
